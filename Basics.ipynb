# Predicting Car Prices

This is a guided project from Dataquest.io. The goal here is to predict the price of some cars based on its attributes.

The Database used can be found at https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data

```python
#Doing the imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
#reading the database file
cars = pd.read_csv("imports-85.data", header=None, names=['symboling','normalized-losses','make','fuel-type','aspiration', 
                                                          'num-of-doors','body-style','drive-wheels','engine-location',
                                                          'wheel-base','length','width','height','curb-weight',
                                                          'engine-type','num-of-cylinders','engine-size','fuel-system',
                                                          'bore','stroke','compression-ratio','horsepower','peak-rpm',
                                                          'city-mpg','highway-mpg','price'], index_col=False)
cars.head(10)
```

```python
#cleaning and normalizing the data
pd.options.display.max_columns = None
cars = cars.replace('?',np.nan)
cars.head(10)
```

```python
cars.dtypes
```

```python
cars['normalized-losses'] = cars['normalized-losses'].astype('float64')
cars['bore'] = cars['bore'].astype('float64')
cars['stroke'] = cars['stroke'].astype('float64')
cars['horsepower'] = cars['horsepower'].astype('float64')
cars['peak-rpm'] = cars['peak-rpm'].astype('float64')
cars['price'] = cars['price'].astype('float64')
```

```python
cars.dtypes
```

```python
cars.info()
```

```python
#Dropping rows where useful information are missing
cars = cars[cars['price'].isnull() == False]
cars.info()
```

```python
cars = cars[cars['peak-rpm'].isnull() == False]
cars.info()
```

```python
cars = cars[cars['bore'].isnull() == False]
cars.info()
```

```python
cars = cars[cars['num-of-doors'].isnull() == False]
cars.info()
```

```python
#For the rows where normalized losses are missing, replacing the values to its median
nl_average_valid = cars[cars['normalized-losses'].isnull()==False]['normalized-losses'].median()
cars = cars.replace(np.nan,nl_average_valid)
cars.head(10)
```
